## 1. Recursión

**Factorial**

* Especificación: 
  - `fac(0) = 1`
  - `fac(n) = n * (n - 1) * ... * 2 * 1`
* Implementación recursiva: 
  - `f(0) = 1`
  - `f(n) = n * f(n - 1)`

```python
# Función factorial de un número natural 
def factorial(n):
   if n == 0:
       return 1
   else:
       return n * factorial(n - 1)

# Ejemplo
num = 9
print("El factorial de", num, "es:", factorial(num))
```

## 2. Relaciones

**Producto cartesiano**

Sean *A* y *B* dos conjuntos cualesquiera su producto cartesiano, se denota *A* $\times$ *B* es el conjunto que contenga todas las parejas ordenadas (*a*, *b*), tales que *a* $\in$ *A* y *b* $\in$ *B*.



```python
# Función que recibe dos conjuntos y regresa su producto cartesiano
def prodCart(setA, setB):
  setAB = set()
  for a in setA:
    for b in setB:
      setAB.add((a, b))
  return setAB

# Ejemplo 
setA = {0, 5, 2, 8}
setB = {1, 3, 4, 7}
setAB = prodCart(setA, setB)
print(setAB)
```

**Ejemplo**

Consideremos el producto cartesiano *A* $\times$ *B* y la **relación** $menor = \{(a, b) | a < b\}$.

*A* = {1, 3, 5, 7, 9, 11, 13, 15}

*B* = {2, 4, 6, 8, 10, 12, 14, 16}


```python
# Función que recibe el producto cartesiano A y B, regresa el conjunto que contenga 
# todas las parejas en las que el primer elemento es menor que el segundo.
def menor(setAB):
  relacion = set()
  for ab in setAB:
    if ab[0] < ab[1]:
      relacion.add(ab)
  return relacion

# Ejemplo
setA = {1, 3, 5, 7}
setB = {2, 4, 6, 8}
setAB = prodCart(setA, setB)
print(menor(setAB))
```

### 2.1 Operaciones con relaciones

#### Unión

$R_1 \cup R_2 = \{(x,y) \, | \, (x,y) \in R_1 \vee (x,y) \in R_2\}$




```python
r1 = {(1, 2), (3, 7), (6, 9)}
r2 = {(4, 5), (8, 1), (3, 2)}

# Función para unir dos relaciones.
print(r1.union(r2))
```

#### Intersección

$R_1 \cap R_2 = \{(x,y) \, | \, (x,y) \in R_1 \wedge (x,y) \in R_2\}$


```python
r1 = {(1, 2), (3, 7), (6, 9)}
r2 = {(6, 9), (1, 2), (3, 2)}

# Operador para realizar la intersección de dos relaciones.
print(r1 & r2)
```

#### Diferencia
$R_1 - \ R_2 = \{(x,y) \, | \, (x,y) \in R_1 \wedge (x,y) \not\in R_2\}$


```python
r1 = {(1, 2), (3, 7), (6, 9)}
r2 = {(6, 9), (1, 2), (3, 2)}

# Operador para realizar la diferencia de dos relaciones.
print(r1 - r2)
```
